### Introduzione e descrizione di CockroachDB

Questa sezione ha lo scopo di descrivere il DBMS scelto per lo svolgimento del progetto:

##### Descrizione del DBMS;

CockroachDB è un sistema di gestione di database **relazionale** distribuito e scalabile, progettato per offrire alta disponibilità, tolleranza ai guasti e consistenza forte. Prende il nome dal suo obiettivo di essere resistente come uno scarafaggio (“cockroach” in inglese), cioè capace di continuare a funzionare anche in presenza di guasti hardware o di rete.

CockroachDB è progettato come un database distribuito nativamente, basato su un’architettura **shared-nothing** in cui ogni nodo del cluster è indipendente e comunica con gli altri tramite rete. Questa struttura permette di ottenere elevata scalabilità, tolleranza ai guasti e performance.

##### Caratteristiche del DBMS;

- **Cluster e Nodi**: un cluster CockroachDB è composto da più nodi (server) che collaborano per fornire una visione unificata del database. Ogni nodo esegue una copia completa del software CockroachDB e può gestire richieste di lettura e scrittura. I nodi condividono responsabilità tramite la distribuzione e replica dei dati.

- **Distribuzione e Replica dei Dati**: i dati nel database sono suddivisi in unità chiamate *range* (partizioni di chiavi, tipicamente da 64 MB). Ogni range è replicato automaticamente su più nodi, tipicamente tre per default, per garantire durabilità e disponibilità. La replicazione è coordinata usando il *protocollo Raft*, un algoritmo di consenso distribuito che assicura che tutte le repliche di un range siano sincronizzate e abbiano una copia coerente.

- **Protocollo Raft**: Raft gestisce la replica e la leadership di ogni range. Tra le repliche, viene eletto un leader che coordina le operazioni di scrittura e assicura che i cambiamenti siano replicati alle altre repliche (*follower*) prima di confermarli al client. In caso di guasti o disconnessioni, Raft elegge un nuovo leader senza interruzioni significative.

- **Transazioni ACID e Consistenza Forte**: CockroachDB supporta transazioni ACID distribuite, garantendo che le operazioni siano atomiche, consistenti, isolate e durevoli anche su più nodi. La consistenza forte è garantita da un sistema di timestamp e da un protocollo di serializzazione ottimistica, che evita conflitti e mantiene l’integrità dei dati.

- **SQL Layer e Storage Layer**: CockroachDB espone un’interfaccia SQL compatibile con PostgreSQL, permettendo agli sviluppatori di usare query SQL familiari. Dietro questa interfaccia, il motore di storage gestisce la suddivisione, la replica e la sincronizzazione dei dati a basso livello.

- **Bilanciamento e Failover**: il sistema monitora continuamente la salute dei nodi e il carico di lavoro. Se un nodo fallisce, i range replicati su quel nodo sono serviti da altre repliche, garantendo la continuità del servizio. Inoltre, i dati possono essere riequilibrati automaticamente tra i nodi per ottimizzare le prestazioni e l’utilizzo delle risorse.

- **Scalabilità Orizzontale**: è possibile aggiungere nodi al cluster senza downtime, e CockroachDB ridistribuisce automaticamente i range per bilanciare carico e spazio. Questa scalabilità orizzontale è ideale per applicazioni con crescita dinamica o carichi variabili.

#### Connessione

Successivamente è necessario connettersi all'istanza di CockroachDB in esecuzione all'interno del Docker container. Per fare ciò, importiamo la libreria psycopg2 (installata nella cella precedente) e specifichiamo a quale istanza di CockroachDB vogliamo connetterci.

In questo esempio, l'istanza è una sola ed è accessibile tramite l'host host.docker.internal. Questo indirizzo permette di raggiungere il container dal sistema host.

Il metodo utilizzato per connettersi al database è *connect* di psycopg2, che prende come parametri:

- **HOST**: Indirizzo o hostname del server CockroachDB (in questo caso host.docker.internal per collegarsi al container)
- **USER**: Username per l'autenticazione (di default root)
- **PASSWORD**: Password per l'autenticazione (vuota se si usa la modalità --insecure)
- **PORTA**: Porta usata dal server CockroachDB (default 26257)
- **NOME DEL DATABASE**: Il database a cui vogliamo connetterci (in questo caso *movr*)

Una volta stabilita la connessione, creiamo un oggetto chiamato *cursor* che ci permette di eseguire query SQL e recuperare i risultati.

`````python
import psycopg2

HOST = "host.docker.internal"  
USER = "root"                 
PASSWORD = ""                 
PORT = 26257                  
DATABASE = "movr"              

conn = psycopg2.connect(
    host=HOST,
    user=USER,
    password=PASSWORD,
    port=PORT,
    database=DATABASE,
    sslmode="disable"  
)

cursor = conn.cursor()
`````

Successivamente vengono definiti due metodi di supporto per effettuare le query SQL che andremo a scrivere nelle celle successive. Entrambi i metodi richiamano la funzione execute del cursore, che esegue la query e permette di recuperare i risultati.

1. **queryCockroachDB(query)**: esegue la query e restituisce tutti i risultati con fetchall()
2. **queryCockroachDB_fetchone(query)**: esegue la query e restituisce la prima riga del risultato con fetchone()

`````python
def queryCockroach(query):
    try:
        cursor.execute(query)
        conn.commit()
        try:
            return cursor.fetchall() 
        except:
            return None
    except Exception as e:
        conn.rollback()
        print("Errore nella query:", e)

def queryCockroach_fetchone(query):
    try:
        cursor.execute(query)
        conn.commit()
        try:
            return cursor.fetchone() 
        except:
            return None
    except Exception as e:
        conn.rollback()
        print("Errore nella query:", e)
`````

### Comparazione con i DBMS visti a lezione

Mostrare le principali differenze con i DB visti a lezione:

- Differenze rispetto a come vengono salvati i dati nel DB in analisi;
- Differenze nella struttura delle query, ad esempio se è più o meno intuitiva rispetto ai linguaggi di interrogazione visti a lezione;
- Analisi sulla documentazione (e.g., più o meno intuitiva e semplice rispetto ai DB visti a lezione).

#### MySQL

- *Salvataggio dei dati* \
  Sia MySQL che CockroachDB sono due RDBMS, la differenza sostanziale è che CockroachDB è un NewSQL, quindi unisce SQL a caratteristiche di DB distribuiti. \
  MySQL organizza i dati in tabelle, righe e colonne, salvandoli tendenzialmente su un unico server. CockroachDB, invece, salva i dati su più nodi in modo automatico, mantenendo così una **consistenza** forte anche in caso di guasti. Grazie al *protocollo Raft* che, ad ogni scrittura e modifica, riesce a garantire che tutti i nodi siano allineati, CockroachDB risolve il problema di **scalabilità orizzontale** presente in MySQL.
- *Struttura delle query* \
  Su query di "livello base" e di bassa complessità i due DBMS sono praticamente identici; le differenze arrivano quando si alza la complessità delle query, rendendo le query CockroachDB più complicate a causa della sua natura distribuita.
- *Documentazione* \
  La documentazione di CockroachDB è moderna, dettagliata e ricca di esempi; ha lo svantaggio di essere un DBMS molto recente (2017), cosa che rende molto più difficile trovare risposte su forum online legate a problemi molto specifici. MySQL, invece, essendo stato rilasciato nel 2001, è molto più popolare, cosa che permette agli utenti di trovare maggiori informazioni online riguardanti ad esso; la documentazione però risulta meno moderna.

#### MongoDB

- *Salvataggio dei dati* \
  Tra CockroachDB e MongoDB c'è una differenza sostanziale dato che il primo è NoSQL e il secondo è un NewSQL. MongoDB salva i dati in documenti BSON (una versione binaria di JSON) e vengono raccolti in collezioni. Hanno la caratteristica comune di supportare la **scalabilità orizzontale**, MongoDB grazie allo *sharding* che suddivide i dati su più nodi e CockroachDB grazie al *protocollo Raft*.
- *Struttura delle query* \
  Le query di MongoDB sono in formato JSON cosa che, per coloro che sono poco pratici del formato, potrebbe causare problemi a causa della poca intuitibilità rispetto al formato SQL. \
  *esempio:* ``` db.utenti.find({ età: { $gt: 18 } }) ```

- *Documentazione* \\
  MongoDB ha una documentazione molto moderna e, essendo molto più popolare, ha anche un grande supporto online della community, con un gran numero di forum, video tutorial su Youtube e Reddit. 

#### Neo4j

- *Salvataggio dei dati* \
  Anche con Neo4j ci sono differenze sostanziali essendo un database a grafo che, quindi, gestisce i dati attraverso l'utilizzo di Nodi (entità) e Relazioni (legame tra entità). Il vantaggio di Neo4j è che le **relazioni sono di prima classe**, cioè che sono parte integrante della struttura dei dati, cosa che permette un accesso molto rapido ai legami tra le entità, anche nel caso di legami complessi.
   
- *Struttura delle query* \
  Neo4j utilizza *Cypher*, un linguaggio dichiarativo e specifico per i grafi, ottimizzato per lavorare con relazioni complesse.\
  *esempio:* ```MATCH (p:Persona {nome: "Simone"})-[:AMICO_DI]->(amico) RETURN amico ```
  `
- *Documentazione* 
  Neo4j ha una buona documentazione moderna con esempi pratici, guide e un'interfaccia grafica (Neo4j Browser) molto utile per visualizzare i grafi.

3. Ottieni i rider che negli ultimi 30 giorni hanno avuto un guadagno maggiore di 100.00 e ordinali in ordine decrescente.

`````python
import time

query = """
SELECT
  u.id,
  u.name,
  COALESCE(SUM(r.revenue), 0) AS total_revenue
FROM public.users u
LEFT JOIN public.rides r ON u.id = r.rider_id
  AND r.start_time >= NOW() - INTERVAL '30 days'
GROUP BY u.id, u.name
HAVING COALESCE(SUM(r.revenue), 0) > 100
ORDER BY total_revenue DESC;
"""

start_time = time.time()

for row in queryCockroach(query):
    print(row)

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Tempo di esecuzione: {elapsed_time:.4f} secondi")
`````

`````python
import time

query = """
SELECT
  u.name,
  ARRAY_AGG(p.code) AS unused_promo_codes
FROM public.users u
LEFT JOIN LATERAL (
  SELECT p.code
  FROM public.promo_codes p
  WHERE p.expiration_time > NOW()
    AND NOT EXISTS (
      SELECT 1
      FROM public.user_promo_codes upc
      WHERE upc.user_id = u.id
        AND upc.code = p.code
    )
  LIMIT 10
) p ON TRUE
GROUP BY u.name
ORDER BY u.name
LIMIT 100;
"""

start_time = time.time()

for row in queryCockroach(query):
    print(row)

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Tempo di esecuzione: {elapsed_time:.4f} secondi")
`````

5. Per ogni utente trova il totale speso nelle corse e il numero di codici promo usati negli ultimi 6 mesi.

`````python
import time

query = """
WITH user_ride_summary AS (
  SELECT
    rider_id,
    COALESCE(SUM(revenue), 0) AS total_spent
  FROM public.rides
  WHERE start_time >= NOW() - INTERVAL '6 months'
  GROUP BY rider_id
),
user_promo_usage AS (
  SELECT
    user_id,
    COUNT(DISTINCT code) AS promo_codes_used
  FROM public.user_promo_codes
  WHERE "timestamp" >= NOW() - INTERVAL '6 months'
  GROUP BY user_id
)
SELECT
  u.id AS user_id,
  u.name,
  COALESCE(urs.total_spent, 0) AS total_spent,
  COALESCE(upu.promo_codes_used, 0) AS promo_codes_used
FROM public.users u
LEFT JOIN user_ride_summary urs ON u.id = urs.rider_id
LEFT JOIN user_promo_usage upu ON u.id = upu.user_id
ORDER BY total_spent DESC;
"""

start_time = time.time()

for row in queryCockroach(query):
    print(row)

end_time = time.time()
elapsed_time = end_time - start_time
print(f"Tempo di esecuzione: {elapsed_time:.4f} secondi")
`````
